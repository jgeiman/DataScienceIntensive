# Data Story 1 - NFL Combine

Each year the National Football League (NFL) holds a combine for a select group of collegiate players from around the country to show off their football skills to prospective teams. It also allows NFL teams to compare players through objective  football-related measures. Football, like most professional sports, is big business and teams invest a lot of time and money in evaluating players. The combine provides metrics on players size, speed, power, and intellect. 

However, critics argue that the combine increases the draft status of "workout warriors" with pedestrian college careers, and unfairly downgrades players with excellent college careers who don't perform well. So let's take a look at some of the metrics measured during the NFL combine and see how they effect a player's draft status. 

****

## Size Matters (sometimes)
NFL players, in many ways, are larger than life. But does size alone translate into improved draft status. For some positions, particularly offensive and defensive lineman, size matters. Below are the top ten heaviest players at the NFL combine from 1999 - 2014. Five of the top ten heaviest players were First Round draft picks in the NFL Draft, but the remaining five heaviest players were either either undrafted, or drafted late. So size does matter, particularly for the offensive tackle position, but clearly more is required than size alone.


```python
%%html
biggest.html
```

To further examine the effect of size on draft status, we can construct a hypothesis test to see if offensive tackles taken in the first round of the NFL draft are statistically larger in weight than offensive tackles overall. The null hypothesis ($H_0$) is that there is no difference in the mean weight of offensive tackles in the first round compared to offensive tackles in general. The alternative hypothesis ($H_A$) is that the mean weight of offensive tackles taken in the first round is greater. The average weight of first round offensive tackles was **319.5 lbs** and the average weight of all offensive tackles was **316.7 lbs**, an average increase of 2.8 lbs. However, this increase in weight was not statistically significant at a significance level of $\alpha=0.05$. This confirms our earlier conclusion that size may be somewhat important for offensive tackles, but clearly more is required than size alone.

****

## Brainiacs
One of the more controversial metrics measured during the NFL Combine is the Wonderlic Cognitive Ability Test, an intelligence test used to measure player's learning and problem solving skills. Below are the ten highest Wonderlic scores at the NFL combine from 1999 - 2014. Interestingly, nine of the top ten players were Quarterbacks. It would seem that intelligence is highly values by NFL teams for the quarterback position. However, only four of the top ten players with the highest Wonderlic scores were selected in the first round of the NFL draft. 


```python
%%html
smartest.html
```

As we did with the size of offensive tackles, we can construct a hypothesis test to see if quarterbacks taken in the first round of the NFL draft score statistically higher on the Wonderlic Test than quarterbacks do overall. The null hypothesis ($H_0$) is that there is no difference in the mean Wonderlic scores of quarterbacks in the first round compared to quarterbacks in general. The alternative hypothesis ($H_A$) is that the mean Wonderlic scores of quarterbacks taken in the first round is greater. The average Wonderlic score of first round quarterbacks was **28** and the average Wonderlic score of all quarterbacks was **26**. This means that a first round quarterback typically scored 2 points higher on the Wonderlic test than all quarterbacks. This increase in the Wonderlic scores of first round quarterbacks was found to be statistically significant at a significance level of $\alpha=0.05$. However, the difference may not be practically significant because we were barely able to reject the null hypthosis (*p-value = 0.0497*). 

So it appears that intelligence will only get you so far in the NFL, and other physical traits are major factors in teams draft decisions. A [2005 study](http://www.thesportjournal.org/article/intelligence-and-football-testing-differentials-collegiate-quarterback-passing-performance-a) found no correlation between a quarterback's Wonderlic score and his passer rating or his salary.

## Raw Power
Strength and power are one of the hallmarks of a great football player. So it follows that the strongest players may have an advantage, and you might expect that strength to translate into success on the football field. One of the measures of strength tested at the combine is the number of repetitions a player can bench press 225 pounds. Below are the top ten players on the bench press at the NFL combine from 1999 - 2014. As with size and intelligence, one position dominates the top players in the bench press. Defensive lineman account for eight of the top ten players in the bench press. This is makes sense, given that their job involves pushing around offensive lineman, which, as we saw earlier, are typically the heaviest players. However, only three of the top ten players in the bench press were taken in the first or second rounds of the NFL draft. So, once again, it appears that more than strength alone is needed to garner an early selection in the NFL draft. 

```python
%%html
strongest.html
```

We can construct a hypothesis test to see if defensive tackles taken in the first round of the NFL draft are statistically stronger (in terms of bench press repetitions) than defensive tackles overall. The null hypothesis ($H_0$) is that there is no difference in the mean number of bench press repetitions of defensive tackles in the first round compared to defensive tackles in general. The alternative hypothesis ($H_A$) is that the mean number of bench press repetitions of defensive tackles taken in the first round is greater. The average number of bench press repetitions of first round defensive tackles was **29.4** and the average number of bench press repetitions of all defensive tackles was **27.5**. On average, first round defensive tackles had almost two more bench press repetitions than defensive tackles overall. This increase in bench press repetitions was found to be statistically significant at a significance level of $\alpha=0.05$.



****

Bench press is only one measure of the strength of NFL players. In his book, *Take Your Eye Off the Ball* Pat Kirwan discusses a simple measure to gauge a players explosiveness, which he calls an Explosion Number. The explosion number is:

EXPLOSION NUMBER = BENCH PRESS + VERTICAL JUMP + BROAD JUMP

This number takes into account both upper and lower body strength. Below is a table of the top ten most explosive players at the NFL combine from 1999-2014. Defensive ends and outside linebackers, which are very similar positions in today's NFL, account for all of the top ten most explosive players. It is interesting how none of the players with the highest bench press repetitions, mostly defensive tackles, also had high explosion numbers. Six of the top ten most explosive players were drafted in the first or second round, however four of the top five were drafted after the second round.

```python
%%html
explosion.html
```

Based on the top ten most explosive players, it may seem likely the explosion number is not that indicative of a player's draft status. To investigate this further, we can construct a hypothesis test to see if outside linebackers and defensive ends taken in the first round of the NFL draft are statistically more explosive than players at these positions overall. The null hypothesis ($H_0$) is that there is no difference in the mean explosion number of first round players compared to players in general at these positions. The alternative hypothesis ($H_A$) is that the mean explosion number of outside linebackers and defensive ends taken in the first round is greater. The average explosion number of first round defensive ends was **167** and the average explosion number of all defensive ends was **162**. The average explosion number of first round outside linebackers was **180** and the average explosion number of all outside linebackers was **174**. The increases in explosion numbers for both defensive ends and outside linebackers was found to be statistically significant at a significance level of $\alpha=0.05$. 

The average explosion number for defensive ends and outside linebackers is plotted below, based on the round in which they were drafted. In general, players taken in later rounds had lower explosion numbers than those taken in earlier rounds. It appears that the explosion number captures some desirable traits in NFL defensive ends and outside linebackers that results in a higher draft status for those that perform well in the these metrics.

Taking this a step further, linear regression can be used to develop a model of the relationship between draft round and explosion number for defensive ends and outside linebackers. Only the slope of the linear regression for outsider linebackers was statistically significant ($\alpha=0.05$). There is a large amount of variation in the explosion number for each round and as a result this model using only draft round accounts for little of the variation in explosion number ($R^2<0.1$).

## Speed Kills
For other positions, while size, strength, and intellect are desired, speed is the primary asset on which players are evaluated. Below is a table of the ten fastest times in the 40 yard dash at the NFL Combine from 1999 - 2014. Wide receivers, cornerbacks, and running backs account for all of the ten fastest times. Only one of the top ten fastest players were drafted later than the fourth round. 

```python
%%html
fastest.html
```

However, as with the explosion numbers, it is unclear whether fast times in the 40 yard dash are correlated with a player's draft status, just by looking at this table. As we've done before with metrics for other positions, we can construct a hypothesis test to see if there statistically significant differences in forty yard dash times for wide receivers and cornerbacks between players taken in the first round compared to all players selected at those positions. The average 40 yd dash time for wide receivers selected in the first round was **4.46 s**, compared to **4.49** for all receivers. Likewise, the average 40 yd dash time for cornerbacks selected in the first round was **4.42 s**, compared to **4.47** for all receivers. Although these differences seem minimal, in both cases the differences in 40 yd dash times were found to be statistically significant at a signficance level of $\alpha=0.05$.

Below is a figure looking at the average 40 yard dash times of cornerbacks and wide receivers by the round in which they were drafted. For these positions, faster times in the 40 yard dash meant higher draft status, with slower players being taken later in the NFL draft, if at all.

# Lessons Learned
We looked at some of the metrics measured during the NFL combine to see how they effect a player's draft status. Some of the lessons learned from this exploration of the NFL combine data include: 

- Relevant metric(s) vary by position, for example:
    - Size (weight) for offensive tackles
    - Wonderlic score for quarterbacks
    - Bench press for defensive tackles
    - Forty yard dash time for wide receivers, cornerbacks, and running backs
    - Vertical leap for running backs
- Some metrics appear to be clearly correlated with draft position.
- First round draft picks have statistically significant differences in important metrics for their positions:
    - 1st Round quarterbacks score higher on the wonderlic test
    - 1st Round defensive tackles bench press more repetitions
    - 1st Round defensive ends and outside linebackers have higher explosion numbers 
    - 1st Round wide receivers and cornerbacks have faster forty yard dash times
- Attempts to model the expected metrics of NFL players by draft round using linear regression found that relationships do exist, however the simple models employed accounted for only a small portion of the variation in these metrics.
- Single metrics may not be the best indicator of draft position. Combinations of metrics may provide improved results.

This analysis also raises several additional questions for further analysis:
1. What metrics, or combination of metrics, should NFL teams use to best evaluate players at each position?
2. What metrics, if any, provide useful predictive measures of a player's future NFL success?

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

combine = pd.read_csv('combine.csv')
```

```python
# remove 2015 data -- incomplete
combine = combine[combine.year != 2015]
```

```python
# listing of data columns
combine.info()
```

```python
# create new column to group like positions into categories
new_positions = {'RB':'RB', 'WR':'WR', 'OLB':'LB', 'FS': 'DB', 'DE':'DL', 'TE':'TE', 'ILB':'LB', 
                 'P':'ST', 'QB':'QB', 'DT':'DL', 'LS':'ST', 'K':'ST', 'FB':'FB',
                 'OG':'OL', 'OT':'OL', 'SS':'DB', 'CB':'DB', 'C':'OL', 
                'NT':'DL', 'OC':'OL'}

combine['pos_cat'] = combine.position.map(new_positions)
pd.unique(combine.pos_cat.values.ravel())
```

```python
# create table with nonzero times
timed40 = combine[combine.fortyyd != 0]

# group times by position and round
fortys = timed40.groupby(['position','round'])[['fortyyd']].mean()
grouped = fortys.unstack('round')

# update column names
grouped.columns = ['Undrafted'] + ['%d' % i for i in range(1,9)]
grouped.sort(axis=1, inplace=True)
del grouped['8']

# plot the results for CB and WR
fig, ax = plt.subplots(1,1)
grouped.loc['CB'].plot(ax=ax, label='CB')
grouped.loc['WR'].plot(ax=ax, label='WR')

ax.set_ylabel('40 yard dash (s)')
ax.set_xlabel('Draft Round')

ax.set_ylim([4.35, 4.6])
ax.legend(loc='best')
fig.savefig('speed_by_round.png', dpi=150)
```

```python
# create a function to map the numeric draft rounds to labels
def round_map():
    rounds = {i:i for i in range(1,8)}
    rounds[0]='Undrafted'
    return rounds
```

```python
# get a table of the 10 fastest players
fastest = timed40.sort('fortyyd')[['name','position','fortyyd', 'round']].head(10)
fastest.columns = ['Name', 'Position', 'Time', 'Round']
fastest['Round'] = fastest.Round.map(round_map())
fastest.set_index('Name', drop=True, inplace=True)
fastest.to_html('fastest.html')
fastest
```

```python
# get a table of the 10 strongest players
strongest = combine[combine['bench']>0]
strongest = strongest.sort('bench', ascending=False)[['name','position','bench', 'round']].head(10)
strongest.columns = ['Name', 'Position', 'Reps', 'Round']
strongest['Round'] = strongest.Round.map(round_map())
strongest.set_index('Name', drop=True, inplace=True)
strongest.to_html('strongest.html')
strongest
```

```python
# get a table of the 10 smartest players
smartest = combine[combine['wonderlic']>0]
smartest = smartest.sort('wonderlic', ascending=False)[['name','position','wonderlic', 'round']].head(10)
smartest.columns = ['Name', 'Position', 'Score', 'Round']
smartest['Round'] = smartest.Round.map(round_map())
smartest.set_index('Name', drop=True, inplace=True)
smartest.to_html('smartest.html')
smartest
```

```python
# get a table of the 10 biggest players (by weight)
biggest = combine.sort('weight', ascending=False)[['name','position','weight', 'round']].head(10)
biggest.columns = ['Name', 'Position', 'Weight', 'Round']
biggest.set_index('Name', drop=True, inplace=True)
biggest['Round'] = biggest.Round.map(round_map())
biggest.to_html('biggest.html')
biggest
```

```python
# get a table of the 10 most explosive players
exp = combine[combine.pos_cat.isin(['OL','DL','LB'])]
exp['explosion'] = exp['bench'] + exp['vertical'] + exp['broad']
exp = exp.sort('explosion', ascending=False)[['name','position','explosion', 'round']].head(10)
exp.columns = ['Name', 'Position', 'Explosion Number', 'Round']
exp['Round'] = exp.Round.map(round_map())

exp.set_index('Name', drop=True, inplace=True)
exp.to_html('explosion.html')
exp
```

```python
# plot the explosion numbers of OLB and DE by draft round
exp = combine[combine.position.isin(['OLB', 'DE'])]
exp = exp[(exp['bench']!=0) & (exp['vertical']!=0) & (exp['broad']!=0)]
exp['explosion'] = exp['bench'] + exp['vertical'] + exp['broad']

expl = exp.groupby(['position','round'])[['explosion']].mean()
grouped = expl.unstack('round')
grouped.columns = ['Undrafted'] + ['%d' % i for i in range(1,9)]
grouped.sort(axis=1, inplace=True)
del grouped['8']

fig, ax = plt.subplots(1,1)
for i in grouped.index.values:
    grouped.loc[i].plot(ax=ax, label=i)

ax.set_ylabel('Explosion Number')
ax.set_xlabel('Draft Round')

ax.legend(loc='best')
fig.savefig('exp_by_round.png', dpi=150)
```

```python
# plot the 40 yd dash and vertical leap of running backs by draft round
rbs = combine[(combine.position=='RB') & (combine.fortyyd > 0) & (combine.vertical > 0)]

grouped = rbs.groupby(['round'])[['vertical', 'fortyyd']].mean()
grouped.reset_index(inplace=True)

rounds = {i:i for i in range(1,8)}
rounds[0]='Undrafted'
rounds[8]='Undrafted'
    
grouped['round'] = grouped['round'].map(rounds)
grouped.sort('round', inplace=True)

ax = grouped.plot(x='round', y='fortyyd', legend=None)
ax2 = ax.twinx()
grouped.plot(x='round', y='vertical', c='r', ax=ax2, legend=None)

ax.set_ylabel('40 yard dash (s)')
ax2.set_ylabel('Vertical Leap (inches)')
ax.set_xlabel('Draft Round')

ax.spines['top'].set_visible(False) 
ax2.spines['top'].set_visible(False) 
ax.xaxis.set_ticks_position('bottom')

ax.spines['right'].set_color('red')
ax.spines['left'].set_color('blue')

ax.tick_params(axis='y', colors='blue')
ax2.tick_params(axis='y', colors='red')

ax.yaxis.label.set_color('blue')
ax2.yaxis.label.set_color('red')

ax.get_figure().savefig('rb_by_round.png', dpi=150)
```

```python
def test_2_means(arr1, arr2, two_sided=True):
    x1 = arr1.mean()
    n1 = len(arr1)
    s1 = arr1.std()

    x2 = arr2.mean()
    n2 = len(arr2)
    s2 = arr2.std()

    x_diff = x1 - x2
    SE = (((s1 ** 2) / n1) + ((s2 ** 2) / n2)) ** 0.5
    z = x_diff / SE
    if two_sided:
        p = stats.norm.sf(abs(z)) * 2
    else:
        p = stats.norm.sf(abs(z))

    return x_diff, SE, z, p
```
